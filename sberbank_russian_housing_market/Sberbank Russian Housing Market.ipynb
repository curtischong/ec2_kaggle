```python
def num_missing(x):
  return sum(x.isnull())

#Applying per column:
print ("Missing values per column:")
print (train.apply(num_missing, axis=0) )#axis=0 defines that function is to be applied on each column
```

```python
train.state = train.state.map(lambda x: 3 if(x == 33) else x)
train.build_year = train.build_year.map(lambda x: 2007 if(x == 20052009) else x)
```

```python
selected_columns = ['full_sq', 'life_sq', 'floor', 'max_floor', 'build_year', 'num_room', 
                    'kitch_sq', 'state', 'price_doc']
corr = train[selected_columns].corr()
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
```
